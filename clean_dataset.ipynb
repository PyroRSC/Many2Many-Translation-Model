```python
import re
import string

datasets = [
    "bantoanon",
    "bicol",
    "cebuano",
    "chavacano",
    "english",
    "filipino-tl",
    "ilokano",
    "ilonggo",
    "ivatan",
    "kinaray-a",
    "manobo",
    "masbatenyo",
    "pampanga",
    "pangasinan",
    "romblomanon",
    "sambal",
    "spanish",
    "tausug",
    "waray",
    "yakan",
    "yami"
]

for data in datasets:
    f = open(f"dataset/raw-txt-files/{data}_cleaned.sentences.txt")
    txt = f.read()

    punctuation_pattern = re.compile(f"[{re.escape(string.punctuation)}]")
    txt = punctuation_pattern.sub("", txt)
    txt = txt.lower()
    
    f_cleaned = open(f"dataset/cleaned-txt/{data}_cleaned.sentences.txt", "w")
    f_cleaned.write(txt)

    f.close()
    f_cleaned.close()
```
