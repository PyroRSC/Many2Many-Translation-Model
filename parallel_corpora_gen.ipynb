```python
import pandas as pd
rgx = r"[!@#$%\^&*\(\)-_=+\[\]\{\}\\~`,\.\/\?;\'\"]"
parallel_corpora = pd.read_csv('dataset/parallel_corpora.csv')
tl_parallel_corporas = [("bicol","bcl"),
    ("chavacano","chv"),
    ("bantoanon","btn"),
    ("cebuano","ceb"),
    # ("english","en"),
    ("ilokano","ilk"),
    ("ilonggo","ilg"),
    ("ivatan","ivt"),
    ("kinaray-a","kry"),
    ("manobo","mbo"),
    ("masbatenyo","mbyo"),
    ("pampanga","pam"),
    ("pangasinan","pgn"),
    ("romblomanon","rblo"),
    ("sambal","sbl"),
    ("spanish","spa"),
    ("tausug","tsg"),
    ("waray","wry"),
    ("yakan","ykn"),
    ("yami","ymi")]

for lang,code in tl_parallel_corporas:
    if(lang == "chavacano" or lang == "bantoanon" or lang == "cebuano" or lang == "bicol"):
        corpora = parallel_corpora[
            (parallel_corpora['language2'] == 'filipino') 
             & (parallel_corpora['language1'] == lang)
        ][['language1_text', 'language2_text']]
        corpora['language1_text'] = corpora['language1_text'].str.replace(rgx,'',regex=True)
        corpora['language2_text'] = corpora['language2_text'].str.replace(rgx,'',regex=True)
        corpora['language1_text'] = corpora['language1_text'].str.lower()
        corpora['language2_text'] = corpora['language2_text'].str.lower()

    else:
        corpora = parallel_corpora[
            (parallel_corpora['language1'] == 'filipino') 
             & (parallel_corpora['language2'] == lang)
        ][['language1_text', 'language2_text']]
        corpora['language1_text'] = corpora['language1_text'].str.replace(rgx,'',regex=True)
        corpora['language2_text'] = corpora['language2_text'].str.replace(rgx,'',regex=True)
        corpora['language1_text'] = corpora['language1_text'].str.lower()
        corpora['language2_text'] = corpora['language2_text'].str.lower()
    corpora.to_csv("dataset/tl-"+code+"-corpora.csv")
```

```python
import re
from pathlib import Path

def clean_file_with_regex(input_path, pattern, replacement, output_path=None):
    """
    Clean a file by applying a single regex substitution.
    
    Args:
        input_path (str or Path): Path to the input file.
        pattern (str): Regex pattern to search for.
        replacement (str): String to replace matches with.
        output_path (str or Path, optional): Output file path. 
            If None, overwrites the input file.
    """
    input_path = Path(input_path)
    
    # Read content
    with open(input_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Apply regex
    cleaned_content = re.sub(pattern, replacement, content)
    
    # Determine output destination
    out_path = Path(output_path) if output_path else input_path
    
    # Write result
    with open(out_path, 'w', encoding='utf-8') as f:
        f.write(cleaned_content)
        
dataset_lang_corpora = [
    ("dataset/cleaned-txt/bicol_cleaned.sentences.txt","bcl"),
    ("dataset/cleaned-txt/chavacano_cleaned.sentences.txt","chv"),
    ("dataset/cleaned-txt/bantoanon_cleaned.sentences.txt","btn"),
    ("dataset/cleaned-txt/cebuano_cleaned.sentences.txt","ceb"),
    ("dataset/cleaned-txt/english_cleaned.sentences.txt","en"),
    ("dataset/cleaned-txt/filipino-tl_cleaned.sentences.txt","tl"),
    ("dataset/cleaned-txt/ilokano_cleaned.sentences.txt","ilk"),
    ("dataset/cleaned-txt/ilonggo_cleaned.sentences.txt","ilg"),
    ("dataset/cleaned-txt/ivatan_cleaned.sentences.txt","ivt"),
    ("dataset/cleaned-txt/kinaray-a_cleaned.sentences.txt","kry"),
    ("dataset/cleaned-txt/manobo_cleaned.sentences.txt","mbo"),
    ("dataset/cleaned-txt/masbatenyo_cleaned.sentences.txt","mbyo"),
    ("dataset/cleaned-txt/pampanga_cleaned.sentences.txt","pam"),
    ("dataset/cleaned-txt/pangasinan_cleaned.sentences.txt","pgn"),
    ("dataset/cleaned-txt/romblomanon_cleaned.sentences.txt","rblo"),
    ("dataset/cleaned-txt/sambal_cleaned.sentences.txt","sbl"),
    ("dataset/cleaned-txt/spanish_cleaned.sentences.txt","spa"),
    ("dataset/cleaned-txt/tausug_cleaned.sentences.txt","tsg"),
    ("dataset/cleaned-txt/waray_cleaned.sentences.txt","wry"),
    ("dataset/cleaned-txt/yakan_cleaned.sentences.txt","ykn"),
    ("dataset/cleaned-txt/yami_cleaned.sentences.txt","ymi")

]
for path,lang in dataset_lang_corpora:
    print(lang)
    clean_file_with_regex(path,rgx,'',path)
# Example usage:
# Remove all digits from 'data.txt' and save to 'clean.txt'
# clean_file_with_regex('data.txt', r'\d+', '', 'clean.txt')

# Or overwrite the original file:
# clean_file_with_regex('data.txt', r'http[s]?://\S+', '')
```
